```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from openai import OpenAI
import os
```

```python
client = OpenAI(api_key=os.getenv('OPENAI_API_KEY'))
```

```python
response = client.responses.create(
    model="gpt-4.1-mini",
    input="who is donald trump bullet points"
)
```

```python
print(response.output_text)
```
